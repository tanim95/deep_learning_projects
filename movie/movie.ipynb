```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movie_df = pd.read_csv('Movie_Id_Titles')
movie_df[movie_df['item_id'] == 50]
movie_df
```

```python
ratings_df = pd.read_csv('u.data',sep = '\t',names = ['user_id','item_id','rating','timestamp'])
ratings_df.head()
```

```python
ratings_df.drop(['timestamp'],axis = 1,inplace = True)
```

```python
ratings_df
```

```python
df = pd.merge(ratings_df,movie_df)
df
```

```python
df.groupby('title')['rating'].describe()
```

```python
df1 = df.groupby('title')['rating'].describe()['mean']
df1
```

```python
df2 = df.groupby('title')['rating'].describe()['count']
df2
```

```python
df3 = pd.concat([df1,df2],axis = 1)
df3 = df3.reset_index()
df3
```

```python
df3.sort_values('count',ascending = False)
```

```python
movietitle_matrix =df.pivot_table(index = 'user_id', columns = 'title', values = 'rating')
# movietitle_matrix = movietitle_matrix.fillna(movietitle_matrix.mean())
movietitle_matrix
```

```python
movietitle_matrix.corr()
```

```python
titanic = movietitle_matrix['Titanic (1997)']
titanic
```

```python
titanic_correlations = pd.DataFrame(movietitle_matrix.corrwith(titanic), columns=['Correlation'])
titanic_correlations
```

```python
titanic_correlations.dropna(inplace=True)
titanic_correlations = titanic_correlations.join(df3['count']).fillna(0)
titanic_correlations
```

```python
titanic_correlations.sort_values('Correlation',ascending = False)
```
